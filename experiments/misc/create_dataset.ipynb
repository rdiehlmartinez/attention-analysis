This notebook's only purpose is to create one joint dataset of bias labels.

```python
import os
```

```python
print(os.getcwd())
```

```python
data_path_raw = "../data/bias_labels.tsv"
data_path_labels = "../data/labeled_bias_data.tsv"
```

```python
f_raw = open(data_path_raw, 'r')
f_labels = open(data_path_labels, 'r')
```

```python
dict_labels = {}
```

```python
next(f_labels)
for line in f_labels:
    parts = line.split('\t') #tab separated 
    if (parts[-1] == 0):
        continue
    dict_labels[parts[1]] = {'epistemological': parts[4], 'framing': parts[5]}
```

```python
dict_raw = {}
```

```python
for line in f_raw: 
    parts = line.strip().split('\t')
    entry_dict = {'pre_tokens':parts[1],
                  'post_tokens':parts[2],
                  'pre':parts[3],
                  'post':parts[4],
                  'pos':parts[5],
                  'rels':parts[6]
                 }
    dict_raw[parts[0]] = entry_dict
```

```python
final_data = {}
```

```python
for entry_id in dict_labels.keys():
    final_data[entry_id] = {**dict_labels[entry_id], **dict_raw[entry_id]}
```

```python
with open("../data/small_bias.tsv", 'w+') as tsvfile:
    fieldnames = ['id', 'pre_tokens', 'post_tokens', 'pos', 'rels', 'epistemological', 'framing']
    
    writer = csv.DictWriter(tsvfile, fieldnames=fieldnames, delimiter='\t')
    writer.writeheader()
    
    for entry_id in final_data.keys():
        writer.writerow({'id': entry_id, 
                         'pre_tokens': final_data[entry_id]['pre_tokens'],
                         'post_tokens': final_data[entry_id]['post_tokens'],
                         'pos': final_data[entry_id]['pos'],
                         'rels': final_data[entry_id]['rels'],
                         'epistemological': final_data[entry_id]['epistemological'],
                         'framing': final_data[entry_id]['framing'],
                        })
```
